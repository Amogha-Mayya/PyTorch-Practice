```python
import torch
```

```python
# example 1 : calculate gradient of y = x^2 , calculate dy/dx
x = torch.tensor(3.0,requires_grad=True)
y = x ** 2
y.backward()
x.grad
```

```python
# example 2 : calcualte gradient of y = x^2 & z = sin(y) , calculate dz/dx
x = torch.tensor(3.0,requires_grad=True)
y = x ** 2
z = torch.sin(y)
z.backward()
x.grad
```

```python
# example 3 : apply autograd on neural network (simple one)
#    z = w*x + b
#    y_pred = sigmoid(z)
#    loss = y_tar log(y_pred) + (1 - y_tar) log(1 - y_pred)

# Calculate dl/dw & dl/db using DERIVATIVE CHAIN RULE
```

```python
x = torch.tensor(6.7,requires_grad=True)
y = torch.tensor(0.0,requires_grad=True)
w = torch.tensor(1.0,requires_grad=True)
b = torch.tensor(0.0,requires_grad=True)
```

```python
# Binary Cross-Entropy Loss for scalar
def binary_cross_entropy_loss(prediction, target):
    epsilon = 1e-8
    prediction = torch.clamp(prediction, epsilon, 1 - epsilon)
    return -(target * torch.log(prediction) + (1 - target) * torch.log(1 - prediction))
```

```python
z = w*x + b
y_pred = torch.sigmoid(z)
loss = binary_cross_entropy_loss(y_pred,y)
```

```python
loss.backward()
w.grad # dl/dw
b.grad # dl/db
```
